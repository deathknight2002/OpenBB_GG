# Biotech Investment Research with OpenBB

This notebook demonstrates how to use OpenBB's biotech extension for pharmaceutical and biotech investment research.

## Features Covered:
- Clinical trial search and analysis
- Company profile research
- Drug pipeline tracking
- Competitive landscape analysis

## Installation

```bash
pip install openbb[biotech]
```

````python
# Import OpenBB
from openbb import obb
import pandas as pd
````

## 1. Clinical Trial Search

Search for clinical trials by disease indication, phase, and status.

````python
# Search for Phase 3 cancer trials that are currently recruiting
cancer_trials = obb.biotech.trials.search(
    query="cancer",
    phase="3",
    status="recruiting",
    provider="clinicaltrials"
)

# Convert to DataFrame for analysis
df_trials = cancer_trials.to_dataframe()
print(f"Found {len(df_trials)} Phase 3 cancer trials")
df_trials.head()
````

````python
# Analyze trials by sponsor
sponsor_counts = df_trials['sponsor'].value_counts().head(10)
print("Top 10 sponsors for Phase 3 cancer trials:")
sponsor_counts
````

## 2. Trial Details

Get detailed information about a specific trial.

````python
# Get details for a specific trial
# Use an NCT ID from the search results above
if len(df_trials) > 0:
    nct_id = df_trials.iloc[0]['nct_id']
    trial_details = obb.biotech.trials.details(
        nct_id=nct_id,
        provider="clinicaltrials"
    )
    
    details_dict = trial_details.to_dict()
    print(f"Trial: {details_dict['results'][0]['title']}")
    print(f"Status: {details_dict['results'][0]['status']}")
    print(f"Enrollment: {details_dict['results'][0]['enrollment']}")
    print(f"\nSummary: {details_dict['results'][0]['brief_summary']}")
````

## 3. Biotech Company Profile

Get comprehensive company profiles for biotech companies.

````python
# Get Moderna's company profile
# Note: This requires FMP provider credentials
try:
    moderna_profile = obb.biotech.profile(
        symbol="MRNA",
        provider="fmp"
    )
    moderna_profile.to_dataframe()
except Exception as e:
    print(f"FMP provider requires API key: {e}")
````

## 4. Drug Pipeline Analysis

Track drug development pipelines for biotech companies.

````python
# Get company pipeline
# Note: This requires FMP provider credentials
try:
    pipeline = obb.biotech.pipeline.company(
        symbol="GILD",
        provider="fmp"
    )
    pipeline.to_dataframe()
except Exception as e:
    print(f"FMP provider requires API key: {e}")
````

## 5. Multi-Indication Analysis

Compare trials across different disease indications.

````python
# Search multiple indications
indications = ["diabetes", "alzheimer", "cardiovascular"]
results = {}

for indication in indications:
    trials = obb.biotech.trials.search(
        query=indication,
        phase="3",
        provider="clinicaltrials"
    )
    results[indication] = len(trials.to_dataframe())

# Create comparison DataFrame
comparison = pd.DataFrame(list(results.items()), columns=['Indication', 'Phase 3 Trials'])
comparison = comparison.sort_values('Phase 3 Trials', ascending=False)
print("\nPhase 3 Trial Count by Indication:")
comparison
````

## 6. Trial Success Analysis

Analyze historical trial success rates.

````python
# Get success rates for oncology
# Note: This requires analytics provider (implementation needed)
try:
    success_rates = obb.biotech.trials.success_rates(
        indication="oncology",
        phase="2",
        provider="analytics"
    )
    success_rates.to_dataframe()
except Exception as e:
    print(f"Analytics provider implementation needed: {e}")
````

## 7. Investment Research Workflow

Complete workflow for researching a biotech investment opportunity.

````python
def research_biotech_company(symbol, indication):
    """
    Complete research workflow for a biotech company.
    
    Args:
        symbol: Company ticker symbol
        indication: Disease indication to focus on
    """
    print(f"\n=== Researching {symbol} in {indication} ===")
    
    # Step 1: Find relevant trials
    print(f"\n1. Searching clinical trials for {indication}...")
    trials = obb.biotech.trials.search(
        query=indication,
        phase="3",
        status="recruiting",
        provider="clinicaltrials"
    )
    df = trials.to_dataframe()
    print(f"   Found {len(df)} Phase 3 trials")
    
    # Step 2: Company profile (requires FMP)
    print(f"\n2. Getting company profile...")
    try:
        profile = obb.biotech.profile(symbol=symbol, provider="fmp")
        print("   Profile retrieved successfully")
    except Exception as e:
        print(f"   FMP provider not configured: {e}")
    
    # Step 3: Pipeline analysis
    print(f"\n3. Analyzing drug pipeline...")
    try:
        pipeline = obb.biotech.pipeline.company(symbol=symbol, provider="fmp")
        print("   Pipeline retrieved successfully")
    except Exception as e:
        print(f"   FMP provider not configured: {e}")
    
    print(f"\n=== Research complete for {symbol} ===")
    return df

# Example: Research Moderna in COVID-19 vaccines
results = research_biotech_company("MRNA", "covid-19")
````
